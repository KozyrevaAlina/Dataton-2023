```python
def ssd(x):
    if 'ТБ' in x:
        tb = float(x.split(' ТБ')[0])
        if 'ГБ' in x:
            gg = float(x.split(' ГБ')[0].split('x')[1])
            return tb*1000*gg
        else:
            return int(tb)*1000
    else:
        return int(x.split(' ГБ')[0])

df['SSDSize_GGb'] = df['SSDSize_GGb'].apply(lambda x: ssd(x) if ((x != np.nan) and (type(x) is str)) else x)
```

```python
df['Battery_hour'].value_counts()
```

```python
def battery(x):
    if 'мин' in x:
        if 'ч' in x:
            minutes  = int(x.split(' мин')[0].split(' ч')[1])
            hour = int(x.split(' мин')[0].split(' ч')[0].split('до ')[1])
            return round(hour + minutes/60, 2)
        else:
            return int(x.split(' мин')[0])
    elif 'час' in x:
        return int(x.split(' час')[0].split(' ')[1])
    else:
        return x

df['Battery_hour'] = df['Battery_hour'].apply(lambda x: battery(x) if ((x != np.nan) and (type(x) is str)) else x)
```
